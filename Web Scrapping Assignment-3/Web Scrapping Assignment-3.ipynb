```python
Datasets_names = []
Data_type = []
Task = []
Attribute_type = []
Num_instance = []
Num_attribute = []
Year = []

for url in dataset_url:
    driver.get(url)
    time.sleep(2)
            
    try:
        types=driver.find_element(By.XPATH,'//p[@class="text-md"][1]')
        Data_type.append(types.text)
    except NoSuchElementException:
        Data_type.append('-')
        
    try:
        task=driver.find_element(By.XPATH,'(//p[@class="text-md"])[3]')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
        
    try: 
        a_types=driver.find_element(By.XPATH,'(//p[@class="text-md"])[4]')
        Attribute_type.append(a_types.text)
    except NoSuchElementException:
        Attribute_type.append('-')
        
    try: 
        num_instance=driver.find_element(By.XPATH,'(//p[@class="text-md"])[5]')
        Num_instance.append(num_instance.text)
    except NoSuchElementException:
        Num_instance.append('-')
        
    try: 
        num_a=driver.find_element(By.XPATH,'(//p[@class="text-md"])[6]')
        Num_attribute.append(num_a.text)
    except NoSuchElementException:
        Num_attribute.append('-')
          
    
    try:
        dataset_name=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Datasets_names.append(dataset_name.text)
    except NoSuchElementException:
        Datasets_names.append('-')
        
        
        
    try: 
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text)
    except NoSuchElementException:
        Year.append('-')         
        
```

```python
df8=pd.DataFrame({'Dataset Name':Datasets_names,'Data Type':Data_type,'Tasks':Task,'Attribute Type':Attribute_type,'Number of Instances':Num_instance,'Number of Attributes':Num_attribute,'Years':Year})
df8
```

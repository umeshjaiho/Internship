```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
header_tags = soup.find('span',class_="mw-headline")
header_tags
```

```python
header_tags.text
```

```python
header = []
for i in soup.find_all('span',class_="mw-headline"):
    header.append(i.text)
    
header    
```

```python
df = pd.DataFrame({'Header Tags':header})
```

```python
df
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
restaurant_name = soup.find('div',class_="restnt-info cursor")
restaurant_name
```

```python
restaurant_name.text
```

```python
restaurant = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    restaurant.append(i.text)
    
restaurant 
```

```python
df = pd.DataFrame({'Restaurant Title':restaurant})
```

```python
df
```

```python
cuisine = soup.find('ul',class_="options")
cuisine
```

```python
cuisine.text
```

```python
cuisines = []
for i in soup.find_all('ul',class_="options"):
    cuisines.append(i.text)
    
cuisines
```

```python
df = pd.DataFrame({'Cuisines':cuisines})
```

```python
df
```

```python
location = soup.find('div',class_="restnt-loc ellipsis")
location
```

```python
location.text
```

```python
Location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
Location
```

```python
df = pd.DataFrame({'Locations':Location})
df
```

```python
rating = soup.find('div',class_="restnt-rating rating-4")
rating
```

```python
Rating = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Rating.append(i.text)
    
Rating
```

```python
df = pd.DataFrame({'Rating':Rating})
df
```

```python
image = soup.find('img',class_="no-img")
image
```

```python
image = []
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
    
image
```

```python
df = pd.DataFrame({'Images':image})
df
```

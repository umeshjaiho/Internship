```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.webdriver.common.keys import Keys
import requests
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in")
```

```python
search=input('Enter the product you want to search:')
```

```python
search1 = driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search1.send_keys(search)
```

```python
searchc=driver.find_element(By.CLASS_NAME,"nav-right")
searchc.click()
```

```python
product_urls=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute("href"))
    next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(3)
```

```python
press=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[3]/div/div/div/div/div[1]/span/a/div/img')
press.click()
```

```python
Brand=[]
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
```

```python
Product_Name=[]
Price=[]
Return=[]
Availability=[]

for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        product_name=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product_Name.append(product_name.text)
    except NoSuchElementException:
        Product_Name.append('-')
        
    try:
        price=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        returns=driver.find_element(By.XPATH,'//div[@class="a-column a-span3 a-text-center icon-container icon-grid-width celwidget"]')
        Return.append(returns.text)
    except NoSuchElementException:
        Return.append('-')
        
    try: 
        availability=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
```

```python
df=pd.DataFrame({'Brand':Brand,'Product Name':Product_Name,'Price':Price,'Return':Return,'Availability':Availability,'Products URL':product_urls})
df
```

```python
#question no 9


driver = webdriver.Chrome()
driver.get("https://www.hostelworld.com/")
```

```python
click_l=driver.find_element(By.CLASS_NAME,"city-card__info-overlay")
click_l.click()
```

```python
urls=[]
url=driver.find_elements(By.XPATH,'//a[@class="property-card-container property-listing-card horizontal"]')
for i in url:
    urls.append(i.get_attribute("href"))
```

```python
overall_review=[]
private=[]
dorms=[]
facilities=[]
escription=[]
```

```python
for url in urls:
    driver.get(url)

    
    try:
        name=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[1]')
        hostel_name.append(name.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    try:
        Distance=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[4]/span[2]')
        distance.append(Distance.text)
    except NoSuchElementException:
        distance.append('-')      
        
    try:
        Rating=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[3]/div/div[1]/span[1]')
        rating.append(Rating.text)
    except NoSuchElementException:
        rating.append('-') 
        
    try:
        t_review=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[3]/div/div[2]')
        total_review.append(t_review.text)
    except NoSuchElementException:
        total_review.append('-')    
        
    try:
        o_review=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[3]/div/div[1]/a/div/div[2]/div[1]')
        overall_review.append(o_review.text)
    except NoSuchElementException:
        overall_review.append('-')   
        
    try:
        privates=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[6]/div/div[1]/div[2]/strong')
        private.append(privates.text)
    except NoSuchElementException:
        private.append('-') 
        
    try:
        dorm=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[2]/div[2]/section/div[2]/a[1]/div[2]/div[6]/div/div[2]/div[2]/strong')
        dorms.append(dorm.text)
    except NoSuchElementException:
        dorms.append('-')   
        
    try:
        description=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[3]/div/div[2]/div/div[2]/div[1]')
        escription.append(description.text)
    except NoSuchElementException:
        escription.append('-')  
        
    try:
        fact=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[3]/main/div[8]/div/ul/li[1]/ul')
        facilities.append(fact.text)
    except NoSuchElementException:
        facilities.append('-')         
```
